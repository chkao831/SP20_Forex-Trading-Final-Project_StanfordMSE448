```python
import os
from pathlib import Path
import pickle
```

```python
temp_store = Path('/home/jingbo/temp_store')

data_types = ['organized_data', 'pickled_data', 
              'candle_10_data', 'candle_30_data', 'candle_60_data', 'candle_300_data', 'candle_600_data']

VALID_PAIRS = ['AUDUSD', 'EURUSD', 'GBPUSD', 'NZDUSD', 'USDCAD', 'USDCHF', 'USDJPY']

ALL_DAYS = [                                                                      "2019-02-01", 
            "2019-02-03", "2019-02-04", "2019-02-05", "2019-02-06", "2019-02-07", "2019-02-08",
            "2019-02-10", "2019-02-11", "2019-02-12", "2019-02-13", "2019-02-14", "2019-02-15",
            "2019-02-17", "2019-02-18", "2019-02-19", "2019-02-20", "2019-02-21", "2019-02-22",
            "2019-02-24", "2019-02-25", "2019-02-26", "2019-02-27", "2019-02-28", "2019-03-01"]
valid_days= [d.replace('-', '') for d in ALL_DAYS]

size_one_folder = len(ALL_DAYS) * 5 * len(VALID_PAIRS)
print(size_one_folder)
```

```python
def examine_folder(folder_name):
    print(folder_name)
    missing = []
    
    all_pairs = sorted(os.listdir(temp_store / folder_name))
    for pair in all_pairs:
        # all_days = sorted(os.listdir(temp_store / folder_name / pair))
        
        for day in valid_days:
            ideal_day_path = temp_store / folder_name / pair / day
            
            if os.path.exists(ideal_day_path):
                all_lps = sorted(os.listdir(ideal_day_path))

                if len(all_lps) != 5:
                    for lp in range(1, 6):
                        ideal_lp_path = temp_store / folder_name / pair / day / str(lp)
                        if os.path.exists(ideal_lp_path):
                            pass
                        else:
                            # missing.append(ideal_lp_path)
                            missing.append((day, pair, lp))
                            print(f'Missing: {ideal_lp_path}')

                else:
                    print(f'{pair} {day} okay')
            else:
                for lp in range(1, 6):
                    missing.append((day, pair, lp))
                
                print(f'Missing: {ideal_day_path}')
    
        print('---')
    
    return missing
```

```python
missing_files = examine_folder('candle_300_data')
print(len(missing_files))
```

```python
pickle.dump(missing_files, open('missing_files.pickle', 'wb'))
```
